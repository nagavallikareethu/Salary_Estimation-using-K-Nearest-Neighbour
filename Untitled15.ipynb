## Salary Estimation | K-nearest neighbour model

```python
import pandas as pd # useful for loading the data set
import numpy as np # to perform array
```

## Choose dataset from local Directory

```python
from google.colab import files
uploaded = files.upload()
```

## Load the dataset

```python
# Get the filename from the uploaded dictionary
filename = list(uploaded.keys())[0]
```

```python
# Now read the file using the correct filename
dataset = pd.read_csv(filename)  #load data set into a variable
dataset
```

## Summarized Dataset

```python
print(dataset.shape) # no.of rows & Columns
print(dataset.head(5)) #top 5 values
```

## Mapping Salary Data to Binary Value

```python
income_set = set(dataset['income'])
dataset['income'] = dataset['income'].map({'<=50K': 0, '>50K': 1}).astype(int)
print(dataset.head(5))
```

## segregate Dataset into X(input/independentvariable)&Y(Output/DependentVariable)

```python
X = dataset.iloc[:, :-1].values
X
```

```python
Y = dataset.iloc[:, -1].values
Y
```

## Splitting Dataset into Train & Test

```python
!pip install scikit-learn
```

```python
#Scikit Learn
from sklearn.model_selection import train_test_split # sklearn - slipt
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.25, random_state = 0)
```

## Feature Scaling
  
   We scale our data to make all the features contribute equally to the result Fit_Transform - fit method is calculating the mean and variance of each of the features present in our data Transform method is transforming all the features using the respective mean and variance, we want our test data to be a completely new and a surprise set for our model

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
X_train
```

## Finding the Best K-value

```python
error = []
from sklearn.neighbors import KNeighborsClassifier #alg
import matplotlib.pyplot as plt #Data visulisation

#calculating error for k values between 1 and the minimum between 40 and the number of training samples
for i in range(1, min(40, len(X_train))): # Ensure n_neighbors is less than or equal to n_samples_fit
    model= KNeighborsClassifier(n_neighbors=i)
    model.fit(X_train, Y_train)
    pred_i = model.predict(X_test)
    error.append(np.mean(pred_i != Y_test))

plt.figure(figsize=(12,6))
plt.plot(range(1, min(40, len(X_train))), error, color='red', linestyle='dashed', marker='o',
         markerfacecolor='blue', markersize=10)
plt.title('Error Rate K Value')
plt.xlabel('K Value')
plt.ylabel('Mean Error')
```

## Training

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors=5, metric='minkowski', p=2)
model.fit(X_train, Y_train)
```

## predicting , Wheather new customer with Age & salary will buy or not

```python
age = int(input("Enter New employees's  Age: "))
edu = int(input("Enter New employees's  Education: "))
cg = int(input("Enter New employees's  Capital-gain: "))
hpw = int(input("Enter New employees's  Hours-per-week: "))
newEmp = [[age, edu, cg, hpw]]
result = model.predict(sc.transform(newEmp))
print(result)

if result == 1:
  print("Employee might got salary above 50k")
else:
  print("Employee might not got Salary above 50k")
```

## prediction for all test Data

```python
y_pred = model.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), Y_test.reshape(len(Y_test),1)),1))
```

## Evaluating Model - Confusion Matrix

```python
from sklearn.metrics import confusion_matrix, accuracy_score
#cm = confusion matrix(y_test,y_pred)

#print("confusion matrix:")
#print(cm)

print("Accuracy of the model: {0}%".format(accuracy_score(Y_test, y_pred)*100))
```
